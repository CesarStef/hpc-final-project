```python
import plotly.express as px
import pandas as pd
#45.095692, 43.178261, 1.917230
# Load the data from the CSV file
data: list[dict] = []
with open('../results/results_strong.txt', 'r') as f:
    lines = f.readlines()
    
    for line in lines:
        if line.startswith('total'):
            pass
        elif line.startswith('execution_time'):
            pass
        elif line.strip() == '':
            pass
        else:
            values = line.split(',')
            data.append({
                'execution_time': float(values[0]),
                'computation_time': float(values[1]),
                'communication_time': float(values[2]),
                'nodes': int(values[3]),
            })

df = pd.DataFrame(data)
print(df)

# # Create a DataFrame from the data

# # Create a line plot for strong scaling
p_tot_strong = px.line(df, x='nodes', y='execution_time',
                         title='Strong Scaling Performance', markers=True)
p_tot_strong.update_layout(xaxis_title='Nodes', yaxis_title='Execution Time (s)')
p_tot_strong.update_layout(xaxis=dict(tickmode='linear', tick0=0, dtick=2, rangemode='tozero'))
p_tot_strong.update_layout(yaxis=dict(tickmode='linear', tick0=0, dtick=5))
p_tot_strong.update_layout(title_x=0.5)
p_tot_strong.update_layout(title=dict(subtitle=dict(text="Grid Size: 15000x15000, Threads: 14, Total Tasks: Nodes * 8")))
#p_tot_strong.update_xaxes({'gridcolor': 'black', 'zerolinecolor': 'blue', 'linecolor': 'white', 'linewidth': 5})

#p_tot_strong.update_xaxes(grdidcolor='lightgray', gridwidth=1)
p_tot_strong.update_layout(plot_bgcolor='white')


p_tot_strong.show()

desired_layout = p_tot_strong.layout

p_comm_strong = px.line(df, x='nodes', y='communication_time',
                         title='Strong Scaling Communication Time', markers=True)
p_comm_strong.update_layout(desired_layout)

p_comm_strong.show()
# # Create a line plot for weak scaling
# fig_weak = px.line(df, x='Total Tasks', y='Execution Time (s)', color='Scaling Type',
#                     title='Weak Scaling Performance', markers=True)
# fig_weak.update_layout(xaxis_title='Total Tasks', yaxis_title='Execution Time (s)')
# fig_weak.write_image(output_file.replace('.png', '_weak.png'))
```
